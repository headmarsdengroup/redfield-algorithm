```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
gs25_aria = np.load('IONQAria_GS_pop_25K1T.npy')
es25_aria = np.load('IONQAria_ES_pop_25K1T.npy')
gs25_simex = np.load('exact_GS_pop_25K1T.npy')
es25_simex = np.load('exact_ES_pop_25K1T.npy')
```

```python
font = {'fontname':'Times New Roman'}
sns.set_palette("rocket", 4)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
plt.rcParams['figure.dpi'] = 360
times = np.load('timesteps.npy')
#plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.plot(times, gs25_aria, linewidth = 0, marker = 'o', alpha = 0.5, label = 'GS IonQ-Aria')
plt.plot(times, gs25_simex, linewidth = 1,  linestyle = 'solid', label = 'GS exact')
plt.plot(times, es25_aria, linewidth = 0, marker = 'o', alpha = 0.5, label = 'ES IonQ-Aria')
plt.plot(times, es25_simex, linewidth = 1,  linestyle = 'dashed',  label = 'ES exact')

plt.ylabel('Population', size = 18, color='#4f4e4e', fontdict=font)
plt.xlabel(r'Time ($\mu s$)', size = 18, color='#4f4e4e', fontdict=font)
plt.xticks(size=14, color='#4f4e4e')
ax.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom=False)
plt.yticks(size=14,color='#4f4e4e')
plt.legend(frameon=True, bbox_to_anchor=(1,1), ncol = 2, fontsize = 8)
sns.despine();
plt.savefig('fig-25K-1T-Aria-data.pdf', format='pdf',bbox_inches="tight")
plt.show()
```
