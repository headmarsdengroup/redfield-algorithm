```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
GS_sim_01T_25K = np.load('IONQ_GS_pop_25K0.1T.npy')
ES_sim_01T_25K = np.load('IONQ_ES_pop_25K0.1T.npy')

GS_sim_1T_25K = np.load('IONQ_GS_pop_25K1T.npy')
ES_sim_1T_25K = np.load('IONQ_ES_pop_25K1T.npy')

GS_sim_5T_25K = np.load('IONQ_GS_pop_25K5T.npy')
ES_sim_5T_25K = np.load('IONQ_ES_pop_25K5T.npy')


GS_ex_01T_25K = np.load('exact_GS_pop_25K0.1T.npy')
ES_ex_01T_25K = np.load('exact_ES_pop_25K0.1T.npy')

GS_ex_1T_25K = np.load('exact_GS_pop_25K1T.npy')
ES_ex_1T_25K = np.load('exact_ES_pop_25K1T.npy')

GS_ex_5T_25K = np.load('exact_GS_pop_25K5T.npy')
ES_ex_5T_25K = np.load('exact_ES_pop_25K5T.npy')

t_list = np.load('tlist.npy')
```

```python
font = {'fontname':'Times New Roman'}
#sns.set_palette("colorblind", 3)
#sns.cubehelix_palette(start=.5, rot=-.5, as_cmap=True, 6)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
plt.rcParams['figure.dpi'] = 360
times = t_list
#plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.plot(times, GS_sim_01T_25K, linewidth = 0, marker = 'o', color = '#3070AD', alpha = 0.5)#, label = 'Ground State 25 K 0.1T')
plt.plot(times, GS_sim_1T_25K, linewidth = 0,marker = 'o', color = '#D39334', alpha = 0.5)#,label = 'Ground State 25 K 1T')
plt.plot(times, GS_sim_5T_25K, linewidth = 0,marker = 'o', color = "#469C76", alpha = 0.5)#,label = 'Ground State 25 K 5T')

plt.plot(times, ES_sim_01T_25K, linewidth = 0,marker = 's', color = '#3070AD', alpha = 0.5)#,label = 'Excited State 25 K 0.1T')
plt.plot(times, ES_sim_1T_25K, linewidth = 0,marker = 's', color = '#D39334', alpha = 0.5)#,label = 'Excited State 25 K 1T')
plt.plot(times, ES_sim_5T_25K, linewidth = 0,marker = 's',color = "#469C76", alpha = 0.5)#,label = 'Excited State 25 K 5T')

gs01 = ax.plot(times, GS_ex_01T_25K, linewidth = 1, linestyle = 'solid', color = '#3070AD', label = '0.1T')
es01 = ax.plot(times, ES_ex_01T_25K, linewidth = 1, linestyle = 'dashed',color = '#3070AD', label = ' ')
gs1 = ax.plot(times, GS_ex_1T_25K, linewidth = 1, linestyle = 'solid', color = '#D39334', label = '1.0T')
es1 = ax.plot(times, ES_ex_1T_25K, linewidth = 1, linestyle = 'dashed', color = '#D39334', label = ' ')
gs5 = ax.plot(times, GS_ex_5T_25K, linewidth = 1, linestyle = 'solid',color = "#469C76", label = '5.0T')
es5 = ax.plot(times, ES_ex_5T_25K, linewidth = 1, linestyle = 'dashed',color = '#469C76', label = ' ')
plt.ylabel('Population', size = 18, fontdict=font)
plt.xlabel('Time ($\mu s$)', size = 18, fontdict=font)
plt.xticks(size=14)
ax.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom=False)

plt.yticks(size=14)
plt.legend(frameon=True, loc = 'best', ncol = 3, bbox_to_anchor=(0.35, 0.7), fontsize = 8)
sns.despine();
plt.savefig('fig-magnetic-field-comparison-25K-v2-plot.pdf', format='pdf',bbox_inches="tight")
plt.show()
```
