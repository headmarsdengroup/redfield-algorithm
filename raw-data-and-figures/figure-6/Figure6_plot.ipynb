```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
exact_magnetization = np.load('mag_list_exact25K1T.npy')
noisy_magnetization = np.load('mag_list_measIONQ25K1T.npy')

t_list = np.load('tlist.npy')
```

```python
font = {'fontname':'Times New Roman'}
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
plt.rcParams['figure.dpi'] = 360
#times = 10**6*np.load('timesteps.npy')
#print(times)
#plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.plot(t_list, noisy_magnetization, linewidth = 0, marker = 'o', color = '#0343DF', alpha = 0.5, label = 'Simulator')
plt.plot(t_list, exact_magnetization, linewidth = 1, linestyle = 'solid', color = '#0343DF', label = 'Exact')
plt.ylabel('Magnetization', size = 18, color='#4f4e4e', fontdict=font)
plt.xlabel(r'Time ($\mu$s)', size = 18, color='#4f4e4e', fontdict=font)
plt.xticks(size=14, color='#4f4e4e')
ax.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom=False)
plt.yticks(size=14,color='#4f4e4e')
plt.legend(frameon=True, loc = 'best', fontsize = 8)
sns.despine();
plt.savefig('fig-magnetization-25K-1T.pdf', format='pdf',bbox_inches="tight")
plt.show()
```
