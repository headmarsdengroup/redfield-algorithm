```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
GS_sim_1T_10K = np.load('IONQ_GS_pop_10K1T.npy')
ES_sim_1T_10K = np.load('IONQ_ES_pop_10K1T.npy')

GS_sim_1T_25K = np.load('IONQ_GS_pop_25K1T.npy')
ES_sim_1T_25K = np.load('IONQ_ES_pop_25K1T.npy')

GS_sim_1T_100K = np.load('IONQ_GS_pop_100K1T.npy')
ES_sim_1T_100K = np.load('IONQ_ES_pop_100K1T.npy')

GS_sim_1T_200K = np.load('IONQ_GS_pop_200K1T.npy')
ES_sim_1T_200K = np.load('IONQ_ES_pop_200K1T.npy')

GS_sim_1T_300K = np.load('IONQ_GS_pop_300K1T.npy')
ES_sim_1T_300K = np.load('IONQ_ES_pop_300K1T.npy')


GS_ex_1T_10K = np.load('exact_GS_pop_10K1T.npy')
ES_ex_1T_10K = np.load('exact_ES_pop_10K1T.npy')

GS_ex_1T_25K = np.load('exact_GS_pop_25K1T.npy')
ES_ex_1T_25K = np.load('exact_ES_pop_25K1T.npy')

GS_ex_1T_100K = np.load('exact_GS_pop_100K1T.npy')
ES_ex_1T_100K = np.load('exact_ES_pop_100K1T.npy')

GS_ex_1T_200K = np.load('exact_GS_pop_200K1T.npy')
ES_ex_1T_200K = np.load('exact_ES_pop_200K1T.npy')

GS_ex_1T_300K = np.load('exact_GS_pop_300K1T.npy')
ES_ex_1T_300K = np.load('exact_ES_pop_300K1T.npy')

t_list = np.load('tlist.npy')
```

```python
font = {'fontname':'Times New Roman'}
sns.set_palette("viridis", 10)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
plt.rcParams['figure.dpi'] = 360
times = 10**6*t_list
#plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.plot(times, GS_ex_1T_10K, linewidth = 1, linestyle = 'solid',  label = '10K')
plt.plot(times, ES_ex_1T_10K, linewidth = 1, linestyle = 'dashed', label = ' ')#, label = 'Excited State 10 K 1T')
plt.plot(times, GS_ex_1T_25K, linewidth = 1, linestyle = 'solid',  label = '25K')
plt.plot(times, ES_ex_1T_25K, linewidth = 1, linestyle = 'dashed', label = ' ')#, label = 'Excited State 25 K 1T')
plt.plot(times, GS_ex_1T_100K, linewidth = 1, linestyle = 'solid', label = '100K')
plt.plot(times, ES_ex_1T_100K, linewidth = 1, linestyle = 'dashed', label = ' ')#, label = 'Excited State 100 K 1T')
plt.plot(times, GS_ex_1T_200K, linewidth = 1, linestyle = 'solid',  label = '200K')
plt.plot(times, ES_ex_1T_200K, linewidth = 1, linestyle = 'dashed', label = ' ')#,  label = 'Excited State 200 K 1T')
plt.plot(times, GS_ex_1T_300K, linewidth = 1, linestyle = 'solid',  label = '300K')
plt.plot(times, ES_ex_1T_300K, linewidth = 1, linestyle = 'dashed', label = ' ')#, label = 'Ground State 300 K 1T')
plt.plot(times, GS_sim_1T_10K, linewidth = 0, marker = 'o', alpha = 0.5)#, label = 'Ground State 10 K 1T')
plt.plot(times, ES_sim_1T_10K, linewidth = 0, marker = 's', alpha = 0.5)#,label = 'Excited State 10 K 1T')
plt.plot(times, GS_sim_1T_25K, linewidth = 0, marker = 'o', alpha = 0.5)#,label = 'Ground State 25 K 1T')
plt.plot(times, ES_sim_1T_25K, linewidth = 0, marker = 's', alpha = 0.5)#, label = 'Excited State 25 K 1T')
plt.plot(times, GS_sim_1T_100K, linewidth = 0, marker = 'o', alpha = 0.5)#,label = 'Ground State 100 K 1T')
plt.plot(times, ES_sim_1T_100K, linewidth = 0, marker = 's', alpha = 0.5)#,label = 'Excited State 100 K 1T')
plt.plot(times, GS_sim_1T_200K, linewidth = 0, marker = 'o', alpha = 0.5)#, label = 'Ground State 200 K 1T')
plt.plot(times, ES_sim_1T_200K, linewidth = 0, marker = 's', alpha = 0.5)#,label = 'Excited State 200 K 1T')
plt.plot(times, GS_sim_1T_300K, linewidth = 0, marker = 'o', alpha = 0.5)# label = 'Ground State 300 K 1T')
plt.plot(times, ES_sim_1T_300K, linewidth = 0, marker = 's', alpha = 0.5)#, label = 'Ground State 300 K 1T')

plt.ylabel('Population', size = 18, fontdict=font)
plt.xlabel(r'Time ($\mu s$)', size = 18, fontdict=font)
plt.xticks(size=14)
ax.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom=False)
plt.yticks(size=14)
plt.legend(frameon=True, ncol = 5, loc = 'best', bbox_to_anchor=(0.15, 0.95), fontsize = 8)
sns.despine();
plt.savefig('fig-temperature-comparison-1T-2025.pdf', format='pdf',bbox_inches="tight")
plt.show()
```
